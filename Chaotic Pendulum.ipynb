## The Chaotic Pendulum

The French mathematician Henri Poincare showed, as early as the end of the nineteenth century, that a simple mechanical system can exhibit complex behavior. The idea that one merely has to specify the initial conditions as accurately as possible to be able, in principle, to precisely predict the motion of a mechanical system from its Newtonian equations of motion is, in practice, taken to the point of absurdity, even by simple models. In general, a system reacts so sensitively to the initial conditions that a minute uncertainty of the initial values will lead to a large uncertainty in the prediction after just a short time. This is true not only for the weather forecast, but also for a simple, externally driven pendulum.

Though this fact has been known for a long time, it took almost 100 years for its significance to be recognized and investigated by the scientific community. Only the computer made it feasible to investigate chaotic motion in detail and -- what is possibly just as important -- display it graphically.

In this section, we want to numerically calculate a simple example: the nonlinear pendulum with frictional force, which is driven by a periodic external force.

```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.animation as animation 
from scipy.integrate import odeint
%matplotlib inline
```

```python
def f(r, t,a,r_,omega_d):
    phi = r[0]
    omega = r[1]
    theta = r[2]
    
    fx = omega 
    fy = -r_*omega - np.sin(phi) + a*np.cos(theta)
    fz = omega_d
    
    return np.array([fx,fy,fz], float)
```

```python
def rk4(f,r,t,h,a,r_,omega_d):
    k1 = h*f(r,t,a,r_,omega_d)
    k2 = h*f(r+0.5*k1,t+0.5*h,a,r_,omega_d)
    k3 = h*f(r+0.5*k2,t+0.5*h,a,r_,omega_d)
    k4 = h*f(r+k3,t+h,a,r_,omega_d)
    
    return (k1+2*k2+2*k3+k4)/6
```

```python
def phase_plot(a,r_,N=10000):
    
    omega_d=2/3
    initial_vals=np.array([np.pi/2, 0.,0],float)
    h = 0.1 #step size
    
    tpoints = []
    for j in range(N):
        tpoints.append(2*np.pi*j/omega_d)

    phi_points = []
    omega_points = []
    theta_points = []
    
    for t in tpoints:
        phi_points.append(initial_vals[0])
        omega_points.append(initial_vals[1])
        theta_points.append(initial_vals[2])
        
        initial_vals += rk4(f,initial_vals,t,h,a,r_,omega_d)
    
        while abs(initial_vals[0]) > np.pi: 
            if initial_vals[0] < -np.pi: 
                initial_vals[0] += 2*np.pi
            if initial_vals[0] > np.pi:
                initial_vals[0] -= 2*np.pi
    
    plt.scatter(phi_points, omega_points, s=0.5)
    plt.xlabel('phi')
    plt.ylabel('w')
    plt.show()
```

```python
phase_plot(0.70,0.25)
```

```python
phase_plot(0.85,0.25)
```

```python
def poincare_section(a):
    steps = 100
    theta_ini = np.pi/2
    omega_ini = 0
    
    r = 0.25
    omega_d = 2/3
    time_step = 2*np.pi/(omega_d*steps)
    
    N = 500
    skip = 100 #throw away first 100
    
    initial_vals = np.array([theta_ini, omega_ini])
    t = np.arange(0.0, N*(2*np.pi)/omega_d, time_step)
    
    def f(y,t):
        g0 = y[1]
        g1 = -r*y[1] - np.sin(y[0]) + a*np.cos(omega_d*t)
        return np.array([g0,g1])
    
    answer = odeint(f, initial_vals, t)
    answer = answer[skip*steps:] #take points each 'steps'
        
    for i, pos in enumerate(answer[:,0]):
        while pos > np.pi:
            pos = pos - (2*np.pi)
        while pos < -np.pi:
            pos = pos + (2*np.pi)
        answer[i,0] = pos
        
    offset = 0
    max_index = (N-skip)*steps - offset
    
    P_phis = []
    P_omegas = []
    
    for j in range(offset,max_index,steps):
    #for j in range(10000,90000,100):
        P_phis.append(answer[j,0])
        P_omegas.append(answer[j,1])
    
    plt.plot(answer[:,0], answer[:,1], 'c,')
    plt.plot(P_phis, P_omegas,'b.')
    plt.xlabel('$\phi$')
    plt.ylabel('$\omega$')
    plt.xlim(-np.pi,np.pi)
    plt.show()

poincare_section(0.70)
```
