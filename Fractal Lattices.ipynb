## Fractal Lattices

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def center(a,b):
    return (a+b)/2
```

```python
np.random.seed(200)

p = np.array(([np.random.randint(0,50),np.random.randint(0,50)],
              [np.random.randint(0,50),np.random.randint(0,50)],
              [np.random.randint(0,50),np.random.randint(0,50)]))
q = np.array([25,1])

points = 10000

xlist = []
ylist = []
for n in range(points):
    r = np.random.randint(0,3) #choose a random point sa P
    q = (q + p[r])/2. #compute midpoint, new q
    xlist.append(q[0]) #store x
    ylist.append(q[1]) #store y

plt.plot(p[0][0],p[0][1], "k")
plt.plot(xlist, ylist,'k.')
plt.axis('off')
plt.show()
```

```python
p = np.array(([0,1], [1,1],[0.5,0]))
q = np.array([0,0])

points = 10000

xlist = []
ylist = []
for n in range(points):
    r = np.random.randint(0,3)
    q = (q + p[r])/2.
    xlist.append(q[0])
    ylist.append(q[1])
plt.plot(xlist, ylist,'k.')
plt.axis('off')
plt.show()
```

```python
p = np.array(([0,0], [1,0],[0.5,0.5]))

def tri(p, count = 0, iterate = 5):
    if count < iterate:
        count += 1
        for i in range(len(p)):
            corner = [p[i]]
            for j in range(len(p)):
                x1 = center(p[i], p[j])
                corner.append(x1)
                z = list(zip(p[i], x1))
                plt.plot(z[0], z[1])
            tri(corner, count)

tri(p)
```
