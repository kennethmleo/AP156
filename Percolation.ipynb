```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.animation as animate
from IPython.display import HTML
%matplotlib inline
```

```python
def jump(i,j):
    directions = [[0,1], [0,-1], [1,0], [-1,0]]
    choice = np.random.randint(0,4)
    
    i += directions[choice][0]
    j += directions[choice][1]
    
    return i, j
```

```python
fig = plt.figure(figsize=(9,9))
ax = fig.add_subplot(111)
ims = []
```

```python
p = 0.59275
L = 25
box = np.zeros([2*L,2*L])
box[L][L] = 1
for i in range(2*L): box[i][-1] = box[i][0] = box[0][i] = box[-1][i] = 2 
    
#0 = unoccupied
#1 = occupied
#2 = vacant

occupiedx = [L]
occupiedy = [L]
x =L
y = L
N = 50000
for i in range(N):
#while 0 in :
    r = np.random.random()
    random = np.random.randint(0,len(occupiedx))
    x,y = jump(occupiedx[random],occupiedy[random])
        
    if box[x][y] == 0:
        if r < p:
            occupiedx.append(x)
            occupiedy.append(y)
            box[x][y] = 1
        elif r>=p: box[x][y] = 2
    
    #im = ax.matshow(box , cmap=cm.gray_r, animated=True)
    #ax.axis('off')
    #ims.append([im])
    
plt.matshow(box , cmap=cm.gray_r, animated=True)
plt.show()
```
