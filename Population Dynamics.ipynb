# Population Dynamics

## Introduction

$logistic\ map$ - iteration of a non-linear function, a parabola that projects the real numbers of the unit interval onto themselves

$deterministic\ chaos$ - well-defined function which yields a sequence of numbers that is practically unpredictable even for a small initial uncertainty

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(r,x):
    return 4*r*x*(1-x)
```

```python
r = 0.87
x0 = 0.65
x = [x0]
N = 41

for i in range(N):
    x.append(f(r,x[i]))

n = np.arange(1,N+1)
plt.plot(n,x[0:N], 'ko', n,x[0:N], 'k--')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Iteration stating from $x_0$ = 0.65 for r = 0.87')
plt.show()
```

```python
x_list = np.linspace(0,1,100)
plt.plot(x_list, x_list, 'k')
plt.plot(x_list, f(r,x_list),label = 'onefold')
plt.plot(x_list, f(r,f(r,x_list)),label = 'twofold')
#plt.plot(x_list, f(r,f(r,f(r,x_list))), label = 'threefold')
plt.plot(x_list, f(r,f(r,f(r,f(r,x_list)))), label = 'fourfold')
plt.xlabel('x')
plt.title('The function $f(x)$ and its two- and fourfold iterations')
plt.legend()
plt.show()
```

```python
def chaos(iterations):
    n = 10000
    r = np.linspace(0,1,n)
    last = 100
    #x = 1e-5
    x = 0.65
    plt.figure(figsize = (9,9))
    for i in range(iterations):
        x = f(r,x)
        if i >= (iterations - last):
            plt.plot(r,x, 'k.', ms = 0.4)


iterations = 1000
chaos(iterations)
plt.xlabel('r', fontsize = 20)
plt.ylabel('x', fontsize = 20)
plt.title('Bifurcation Diagram of the logistic map', fontsize = 25)
plt.savefig('bifurcation.png', bbox_inches = 'tight')
plt.show()
```

```python
def histogram(x=0.3, r = 0.934):
    
    iterations = 10000
    xn = []

    for i in range(iterations):
        x = f(r,x)
        xn.append(x)
    
    # Complete the following line
    plt.figure(figsize=(6,6))
    plt.hist(xn, bins=100)
    plt.xlim(xmin=0)
    # add axis labels
    plt.xlabel('$n$')
    plt.ylabel('frequency')
    plt.title('Frequency distribution of the $x_n$ values')
    plt.show()

histogram()
```
